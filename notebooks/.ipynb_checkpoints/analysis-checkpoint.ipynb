# Retail Sales & Profitability Analysis

SQL + Python analysis of 9,994 retail orders (2014-2017) from the Superstore dataset.
Goal: find where the money is, where it's leaking, and whether discounting helps or hurts.

**Tools:** SQLite (via Python's `sqlite3`), pandas, matplotlib.
**Data:** `data/superstore.csv` — downloaded from Kaggle ([Superstore Dataset](https://www.kaggle.com/datasets/vivek468/superstore-dataset-final)).

## 1. Load the data

The raw CSV is Latin-1 encoded, not UTF-8 — typical of Excel exports on Windows. Trying `utf-8` first throws a decode error; `latin-1` works.

```python
import sqlite3
import re
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/superstore.csv", encoding="latin-1")
print(df.shape)
df.head()
```

## 2. Clean it

Dates come in as US-format strings (`M/D/YYYY`). If you don't parse them before doing
date math, pandas/SQLite will sort `"1/1/2017"` before `"11/8/2016"` alphabetically —
that silently gives you the wrong date range. Parse first, always.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%m/%d/%Y")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%m/%d/%Y")

print("Fully duplicated rows:", df.duplicated().sum())
print("Duplicate Order+Product lines:", df.duplicated(subset=["Order ID", "Product ID"]).sum())
print("Date range:", df["Order Date"].min().date(), "to", df["Order Date"].max().date())
nulls = df.isnull().sum()
print("Nulls per column:", "none" if nulls.sum() == 0 else nulls[nulls > 0])
```

```python
# SQLite has no native DATE type — its date functions (strftime, julianday)
# expect ISO format (YYYY-MM-DD), not US format. Keep both.
df["order_date_iso"] = df["Order Date"].dt.strftime("%Y-%m-%d")
df["ship_date_iso"] = df["Ship Date"].dt.strftime("%Y-%m-%d")
```

## 3. Load into SQLite and apply the schema

`sql/schema.sql` creates:
- `orders_enriched` — a VIEW that precomputes profit margin % once, so every query below just references it
- `discount_tier` — a small reference table (No discount / Low / Medium / High) used later for a range JOIN

```python
conn = sqlite3.connect(":memory:")
df.to_sql("orders", conn, index=False, if_exists="replace")

with open("../sql/schema.sql") as f:
    conn.executescript(f.read())
print("schema applied")
```

## 4. Run each business question

`sql/queries.sql` has 10 labeled queries (`-- Q1.` through `-- Q10.`). We split the file
on those labels and run each one separately so we can inspect results one at a time
below, instead of just dumping everything to CSV.

```python
with open("../sql/queries.sql") as f:
    sql_text = f.read()

blocks = re.split(r"(?=-- Q\d+\.)", sql_text)
blocks = [b.strip() for b in blocks if b.strip().startswith("-- Q")]

results = {}
for block in blocks:
    qid = re.match(r"-- (Q\d+)\.", block).group(1)
    sql_lines = [l for l in block.splitlines() if not l.strip().startswith("--")]
    sql = "\n".join(sql_lines).strip()
    results[qid] = pd.read_sql_query(sql, conn)

print(f"Ran {len(results)} queries:", list(results.keys()))
```

### Q1 — Which Category / Sub-Category combos make money, and which lose it?

Sorted ascending on purpose: the losses show up first.

```python
results["Q1"].sort_values("total_profit").head(8)
```

```python
q1 = results["Q1"].sort_values("total_profit")
colors = ["#C44E52" if v < 0 else "#55A868" for v in q1["total_profit"]]
plt.figure(figsize=(9, 6))
plt.barh(q1["Sub-Category"], q1["total_profit"], color=colors)
plt.title("Total Profit by Sub-Category (red = losing money)")
plt.xlabel("Total Profit (USD)")
plt.tight_layout()
plt.savefig("../visuals/profit_by_subcategory.png", dpi=150)
plt.show()
```

**Finding:** Furniture Tables lose -\$17,725 on \$206,966 of sales (-14.8% margin) — the single
biggest profit leak in the business. Bookcases lose another -\$3,473.

### Q2 — Month-over-month sales trend

Uses the window function `LAG()` to grab the previous row's value without a self-join —
this is the standard way to compute period-over-period growth in SQL.

```python
q2 = results["Q2"].copy()
q2["period"] = q2["order_year"].astype(str) + "-" + q2["order_month"].astype(str).str.zfill(2)
plt.figure(figsize=(11, 4.5))
plt.plot(q2["period"], q2["monthly_sales"], marker="o", markersize=3, color="#4C72B0")
plt.xticks(rotation=90, fontsize=6)
plt.title("Monthly Sales, 2014-2017")
plt.ylabel("Sales (USD)")
plt.tight_layout()
plt.savefig("../visuals/monthly_sales_trend.png", dpi=150)
plt.show()
```

### Q4 — Does discounting kill profit?

A JOIN against `discount_tier` on a **range condition** (`BETWEEN`), not an equality —
a step up from a normal ID-to-ID join.

```python
results["Q4"]
```

```python
q4 = results["Q4"]
colors = ["#C44E52" if v < 0 else "#55A868" for v in q4["avg_margin_pct"]]
plt.figure(figsize=(6, 4.5))
plt.bar(q4["tier_label"], q4["avg_margin_pct"], color=colors)
plt.axhline(0, color="black", linewidth=0.8)
plt.title("Average Profit Margin % by Discount Tier")
plt.ylabel("Avg Profit Margin (%)")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig("../visuals/margin_by_discount_tier.png", dpi=150)
plt.show()
```

**Finding:** margin goes from +34.0% (no discount) to **-108.9%** (41%+ discount). Past
roughly 20% off, orders are routinely sold at a loss bigger than the discount itself.
This is the most actionable finding in the whole dataset — a discount-approval policy
problem, not a "sell more" problem.

### Q6 — The 10 worst products by total profit

```python
results["Q6"]
```

**Finding:** the single worst product is a Cubify CubeX 3D printer, -\$8,880 lost on

### Q8 — Top 3 Sub-Categories by profit, within each Region

`RANK() OVER (PARTITION BY Region ...)` wrapped in a CTE, because you can't filter on a
window function in the same SELECT that defines it — the CTE is what makes the `WHERE
profit_rank <= 3` afterward possible.

```python
results["Q8"]
```

### Q9 — Year-over-year growth by Category

Same `LAG()` idea as Q2, but partitioned by Category — each year is compared to *that
category's own* prior year, not just the row above it.

```python
q9 = results["Q9"].dropna(subset=["yoy_growth_pct"])
plt.figure(figsize=(7, 4.5))
for cat in q9["Category"].unique():
    sub = q9[q9["Category"] == cat]
    plt.plot(sub["order_year"], sub["yoy_growth_pct"], marker="o", label=cat)
plt.axhline(0, color="black", linewidth=0.8)
plt.title("Year-over-Year Sales Growth by Category")
plt.ylabel("YoY Growth (%)")
plt.legend()
plt.tight_layout()
plt.savefig("../visuals/yoy_growth_by_category.png", dpi=150)
plt.show()
```

**Finding:** Technology and Office Supplies grew 20-39% YoY in 2016-2017; Furniture growth
slowed to +8.3% in 2017 after +16.7% in 2016 — and Furniture is also where the profit
leak (Tables, Bookcases) is concentrated. Growing a money-losing category isn't
automatically good news.

## Summary

1. **Furniture Tables (-\$17,725) and Bookcases (-\$3,473) are the biggest profit leaks.**
2. **Discounts past ~20% turn profitable orders into losses** — 41%+ discount averages -108.9% margin.
3. **The worst individual products are high-ticket, high-discount items** (3D printers, specialty machines), not cheap goods.
4. **A subtlety worth remembering:** Machines shows a *negative average margin per order* but *positive total profit* — a few large high-margin sales can outweigh many small losing ones. Average-of-percentages and percentage-of-totals are not the same number; check which one you're looking at.
5. **Growth is concentrated in Technology and Office Supplies**, not in Furniture, the category actively losing money.
